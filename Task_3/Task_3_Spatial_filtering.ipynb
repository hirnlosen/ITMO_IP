# Task 3: Spatial filtering

```python
# Importing required modules
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Defining path to images
images_path = "resources/images/"
```

### Exercise 1

```python
# Reading image with OpenCV
image_1_name = "saint-petersbourg-1936035_1280.jpg"
image_1_bgr = cv2.imread(images_path + image_1_name)

# Converting image from bgr to rgb format and displaying image on the screen
image_1_rgb = cv2.cvtColor(image_1_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_1_rgb)
plt.show()

# Converting image to grayscale and displaying image on the screen
image_1_grayscale = cv2.cvtColor(image_1_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(image_1_grayscale, cmap='gray')
plt.show()
```

```python
# Displaying histogram for number of pixels correspoding to particular intensity
grayscale_hist = plt.hist(image_1_grayscale.ravel(), bins = 256)
plt.xlabel('Intensity $n$')
plt.ylabel('Number $h(n)$')
plt.show()

# Searching for intensity value with the biggest amount of pixels
max_pixels = 0
intensity_index = 0
for i in range(0, len(grayscale_hist[0])):
    if grayscale_hist[0][i] > max_pixels:
        max_pixels = grayscale_hist[0][i]
        intensity_index = i
print("Intensity value with the biggest amount of pixels is: {}".format(grayscale_hist[1][intensity_index]))
print("Number of pixels correspoding to this intensity is: {}".format(grayscale_hist[0][intensity_index]))
```

### Exercise 2

```python
# Reading image with OpenCV
image_2_name = "russia-2396022_1280.jpg"
image_2_bgr = cv2.imread(images_path + image_2_name)

# Converting image from bgr to rgb format and displaying image on the screen
image_2_rgb = cv2.cvtColor(image_2_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_2_rgb)
plt.show()
```

```python
# Displaying histogram for frequency of pixels correspoding to particular intensity
red_hist = plt.hist(image_2_rgb[:, :, 0].ravel(), bins = 256, color = 'Red', alpha = 0.5, density=True)
green_hist = plt.hist(image_2_rgb[:, :, 1].ravel(), bins = 256, color = 'Green', alpha = 0.5, density=True)
blue_hist = plt.hist(image_2_rgb[:, :, 2].ravel(), bins = 256, color = 'Blue', alpha = 0.5, density=True)
plt.legend(['Red channel (R)', 'Green channel (G)', 'Blue channel (B)'])
plt.xlabel('Intensity $n$')
plt.ylabel('Frequency')
plt.show()

# Searching for maximum density value among three channels
max_density_red = 0
max_density_green = 0
max_density_blue = 0
for i in range(0, len(red_hist[0])):
    if red_hist[0][i] > max_density_red:
        max_density_red = red_hist[0][i]
    if green_hist[0][i] > max_density_green:
        max_density_green = green_hist[0][i]
    if blue_hist[0][i] > max_density_blue:
        max_density_blue = blue_hist[0][i]
print("Maximum density value among all channels is: {}".format(max([max_density_red, max_density_green, max_density_blue])))
```

### Exercise 3

```python
# Reading image with OpenCV
image_3_name = "monastery-3714287_1280.jpg"
image_3_bgr = cv2.imread(images_path + image_3_name)

# Converting image from bgr to rgb format and displaying image on the screen
image_3_rgb = cv2.cvtColor(image_3_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_3_rgb)
plt.show()
```

```python
# Normalizing image intensity
image_3_rgb_norm = image_3_rgb.copy()
image_3_rgb_norm[:, :, 0] = (image_3_rgb[:, :, 0] - np.min(image_3_rgb[:, :, 0]))/(np.max(image_3_rgb[:, :, 0]) - np.min(image_3_rgb[:, :, 0]))*255
image_3_rgb_norm[:, :, 1] = (image_3_rgb[:, :, 1] - np.min(image_3_rgb[:, :, 1]))/(np.max(image_3_rgb[:, :, 1]) - np.min(image_3_rgb[:, :, 1]))*255
image_3_rgb_norm[:, :, 2] = (image_3_rgb[:, :, 2] - np.min(image_3_rgb[:, :, 2]))/(np.max(image_3_rgb[:, :, 2]) - np.min(image_3_rgb[:, :, 2]))*255

# Displaying normalized image
plt.imshow(image_3_rgb_norm)
plt.show()

# Printing values for all channels at pixel (424,637)
print("For pixel (424,637) R,G,B channels values are: {}".format(image_3_rgb_norm[424,637]))
```

### Exercise 4

```python
# Reading image with OpenCV
image_4_name = "st-petersburg-3393467_1280.jpg"
image_4_bgr = cv2.imread(images_path + image_4_name)

# Converting image from bgr to rgb format and displaying image on the screen
image_4_rgb = cv2.cvtColor(image_4_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(image_4_rgb)
plt.show()
```

```python
# Applying Gaussian blur to the image
image_4_rgb_blurred = cv2.GaussianBlur(image_4_rgb, (5,5), 0)
plt.imshow(image_4_rgb_blurred)
plt.show()
```

```python
# Processing image difference
image_4_difference = image_4_rgb - image_4_rgb_blurred
plt.imshow(image_4_difference)
plt.show()
```
