```python
# Importing required modules
import cv2
import matplotlib.pyplot as plt
```

```python
# Defining path to images
images_path = "resources/images/"
```

### Exercise 1

```python
# Reading image with OpenCV
image_1_name = "flower_1.jpg"
image_1 = cv2.imread(images_path + image_1_name)

# Getting basic image info
height_1, width_1, color_channels_1 = image_1.shape
print("Image 1 resolution is {}x{}".format(width_1, height_1))

# Getting image color values at 509 row and 980 column
b, g, r = image_1[509][980]
print("Image 1 red channel value: {}, green channel value: {}, blue channel value: {}".format(r, g, b))
```

### Exercise 2

```python
# Reading image with OpenCV
image_1_name = "flower_1.jpg"
image_1 = cv2.imread(images_path + image_1_name)

# Converting image from bgr to rgb format
image_1 = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)

# Croping image
image_1_cropped = image_1[110:879, 720:1279]

# Displaying image
plt.imshow(image_1_cropped)
plt.show()
```

### Exercise 3

```python
# Reading image with OpenCV
image_2_name = "flower_2.jpg"
image_2 = cv2.imread(images_path + image_2_name)

# Converting image from bgr to rgb format
image_2 = cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB)

# Getting basic image info
height_2, width_2, color_channels_2 = image_2.shape
#print("Image 2 resolution is {}x{}".format(width_2, height_2))

# Calculating proportionality coefficient
k = float(width_2) / height_2
print("Aspect ratio of image is: {}".format(k))

# Compressing image proportionally
new_width = 620
new_height = int(new_width / k)
#print("New image height is {}".format(new_height))
dim = (new_width, new_height)
image_2_resized = cv2.resize(image_2, dim, interpolation = cv2.INTER_AREA)
plt.imshow(image_2_resized)
plt.show()

# Getting resized image info
height_2, width_2, color_channels_2 = image_2_resized.shape
print("Resized image 2 resolution is {}x{}".format(width_2, height_2))

# Saving resized image to the file
cv2.imwrite(images_path + "flower_2_resized.jpg", cv2.cvtColor(image_2_resized, cv2.COLOR_RGB2BGR))

# Increasing intensity of r channel for 33%
image_2_increased_intensity = image_2_resized.copy()
image_2_increased_intensity[:, :, 0] = image_2_increased_intensity[:, :, 0] * 1.33
plt.imshow(image_2_increased_intensity)
plt.show()

# Getting image color values at 242 row and 328 column
r, g, b = image_2_increased_intensity[242][328]
print("Image 2 (after increase of r channel) red channel value: {}, green channel value: {}, blue channel value: {}".format(r, g, b))
```
